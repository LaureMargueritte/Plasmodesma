```python
def renorm2D(Z, mu, sigma):
    "renormalise un histogramme"
    lZZ = np.log(Z)
    lZZ = lZZ[np.nonzero(lZZ)]
    lmu = lZZ.mean()
    lsigma = lZZ.std()
    print (lmu,lsigma)
    ZZ = np.log(Z)
    line,col = ZZ.shape
#    for i in range(line):
#        for j in range(col):
#            ZZ[i,j] = (ZZ[i,j]-lmu)*sigma/lsigma + mu
    ZZ = (ZZ-lmu)*sigma/lsigma + mu
    ZZ = np.where(Z==1, 0, ZZ)
    return np.exp(ZZ)
```
